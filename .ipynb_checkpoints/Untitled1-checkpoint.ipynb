```python
Data_Train = pd.read_excel("train.xlsx", engine="openpyxl")
Data_Tast  = pd.read_excel("test.xlsx", engine="openpyxl")
```

```python
Data_Train = Data_Train.sort_values(by='dt')
Data_Tast = Data_Tast.sort_values(by='dt')
```

```python
Data_Train['Rol_Mean_Цена на арматуру'] = Data_Train['Цена на арматуру'].rolling(window=12).mean()
Data_Train['Rolling_Std'] = Data_Train['Цена на арматуру'].rolling(window=12).std()
```

```python
Data_Train['Lag_1'] = Data_Train['Цена на арматуру'].shift(1)
Data_Train['Lag_2'] = Data_Train['Цена на арматуру'].shift(2)
```

```python
Data_Tast['Rol_Mean_Цена на арматуру'] = Data_Train['Цена на арматуру'].rolling(window=12).mean()
Data_Tast['Rolling_Std'] = Data_Train['Цена на арматуру'].rolling(window=12).std()
```

```python
Data_Tast['Lag_1'] = Data_Train['Цена на арматуру'].shift(1)
Data_Tast['Lag_2'] = Data_Train['Цена на арматуру'].shift(2)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Data_Train = Data_Train.dropna()
```

```python
Data_Tast = Data_Tast.dropna()
```

```python
Data_Tast.info() 
Data_Train.info()
```

```python
columns_to_scale = ['Цена на арматуру', 'Rol_Mean_Цена на арматуру', 'Rolling_Std', 'Lag_1', 'Lag_2']

# Создаем scaler и применяем его к обучающим данным
scaler = MinMaxScaler(feature_range=(0, 1))
Data_Train[columns_to_scale] = scaler.fit_transform(Data_Train[columns_to_scale])

# Применяем тот же scaler к тестовым данным
Data_Tast[columns_to_scale] = scaler.transform(Data_Tast[columns_to_scale])
```

```python
import numpy as np
def create_sequences(data, target_column, time_steps=1):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data.iloc[i:(i + time_steps), 1:].values)  # Все колонки, кроме 'dt'
        y.append(data.iloc[i + time_steps, data.columns.get_loc(target_column)])
    return np.array(X), np.array(y)

# Задаем количество временных шагов
time_steps = 10

# Создаем последовательности для обучающих данных
X_train, y_train = create_sequences(Data_Train, target_column='Цена на арматуру', time_steps=time_steps)

# Создаем последовательности для тестовых данных
X_test, y_test = create_sequences(Data_Tast, target_column='Цена на арматуру', time_steps=time_steps)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Создаем модель
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_steps, X_train.shape[2])))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(1))

# Компилируем модель
model.compile(optimizer='adam', loss='mean_squared_error')

# Обучаем модель
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Прогнозирование на тестовых данных
predicted = model.predict(X_test)

# Обратное преобразование данных
predicted = scaler.inverse_transform(np.concatenate((predicted, np.zeros((predicted.shape[0], 4))), axis=1))[:, 0]
y_test = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros((y_test.shape[0], 4))), axis=1))[:, 0]

# Визуализация результатов
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(y_test, label='Фактические значения')
plt.plot(predicted, label='Прогнозируемые значения')
plt.legend()
plt.show()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Загрузка данных
train_data = pd.read_excel('train.xlsx')
test_data = pd.read_excel('test.xlsx')

# Преобразование дат в числовой формат (количество дней с начальной даты)
train_data['dt'] = pd.to_datetime(train_data['dt'])
train_data['days'] = (train_data['dt'] - train_data['dt'].min()).dt.days

test_data['dt'] = pd.to_datetime(test_data['dt'])
test_data['days'] = (test_data['dt'] - test_data['dt'].min()).dt.days

# Разделение данных на признаки (X) и целевую переменную (y)
X_train = train_data[['days']]
y_train = train_data['Цена на арматуру']

X_test = test_data[['days']]
y_test = test_data['Цена на арматуру']

# Создание и обучение модели
model = LinearRegression()
model.fit(X_train, y_train)

# Прогнозирование на тестовых данных
y_pred = model.predict(X_test)

# Оценка модели
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R² Score: {r2}')

# Визуализация результатов
plt.scatter(X_test, y_test, color='blue', label='Фактические значения')
plt.plot(X_test, y_pred, color='red', label='Прогнозируемые значения')
plt.xlabel('Дни')
plt.ylabel('Цена на арматуру')
plt.legend()
plt.show()
```
