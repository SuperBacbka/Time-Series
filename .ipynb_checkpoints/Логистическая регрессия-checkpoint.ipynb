Попробуем подобрать Гипер параметры для последней модели c помощью RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
```

Гиперпараметры для случайного перебора

```python
param_dist = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 200, 300],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'reg_alpha': np.linspace(0, 1, 5),
    'reg_lambda': np.linspace(0, 1, 5)
}
```

```python
Log_Model5 = XGBClassifier(objective='binary:logistic', eval_metric="auc", random_state=42)
```

Настройка RandomizedSearchCV

```python
random_search = RandomizedSearchCV(
    estimator=Log_Model5,
    param_distributions=param_dist,
    n_iter=50,          # Количество случайных комбинаций
    scoring='roc_auc',  # Оценка по метрике ROC-AUC
    cv=3,               # Кросс-валидация с 3 фолдами
    verbose=2,          # Показывать процесс поиска
    n_jobs=-1,          # Использовать все процессоры
    random_state=42     # Для воспроизводимости
)
```

```python
random_search.fit(scaler_X_train, y_train)
```

```python
print("Лучшие параметры:", random_search.best_params_)
print("Лучший ROC-AUC:", random_search.best_score_)
```

```python
Final_Model = XGBClassifier(
    max_depth=7,#глубина дерева
    learning_rate=0.1,#Скорость обучения
    n_estimators=300,#количество итераций бустинга
    objective='binary:logistic', # задача бинарной классификации, где результатом является вероятность (логистическая регрессия).
    subsample=1.0,#Доля случайно выбранных обучающих образцов для построения каждого дерева.
    colsample_bytree=1.0, #Доля случайно выбранных признаков для построения каждого дерева
    reg_alpha=1.0,#т штраф за абсолютные значения весов листьев,
    reg_lambda=0.25, #Добавляет штраф за квадраты значений весов листьев
    random_state=42,#Устанавливает случайное начальное значение для обеспечения воспроизводимости результатов.
    eval_metric="auc"#Определяет метрику оценки для отслеживания производительности модели во время обучения.
)
```
